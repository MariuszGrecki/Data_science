# Mnożenie macierzy

Aby pomnozyć dwie macierze liczba kolumn w lewej macierzy musi się zgadzać z liczbą wierszy prawej macierzy.

image.png

```python
X = np.array([[5,3], [3,9]])
Y = np.array([[1], [-1]])
print(X)
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
Z.shape
```

```python
X = np.array([[2, -1, 3],
              [3, 1, 0]])

Y = np.array([[2, 1, -1],
              [0, -1, 2],
              [3, 2, 0]])

print(X, '\n')
print(Y)
```

```python
Z = np.dot(X, Y)
Z
```

```python
A = np.array([[2,4], [1,-1]])
print(A)
```

```python
B = np.array([[10], [-1]])
print(B)
```

```python
A_inv = np.linalg.inv(A)
```

```python
X = np.dot(A_inv, B)
X
```
