# Obliczanie nowych kolumn

```python
df = fetch_financial_data('UBER')
df.info()
```

```python
#obliczanie nowej kolumny dla której bedzie srednia z open i close
df['Average'] = (df.Open + df.Close) / 2
df.head(5)
```

```python
#sortowanie po indeksie
df.sort_index()
```

```python
df.head(3)
```

```python
df = df.sort_index()
```

```python
df.head(5)
```

```python
#obliczanie dziennej zmiany procentowej
df.Close.shift(1) # metoda przesuwa dane o 1 wiersz
```

```python
df[['Close']].shift(2)
```

```python
df['Close_shift'] = df.Close.shift(1)
```

```python
df.head(3)
```

```python
#dzienna zmiana wyrażona w %
df['Daily_Change'] = df.Close / df.Close_shift - 1
df.head()
```

```python
#usuwamy kolumne
del df['Daily Change']
```
